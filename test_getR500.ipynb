```python
%load_ext autoreload
%autoreload 2
from astropy import units as u
from astropy import constants as const
import numpy as np
from matplotlib import pyplot as plt

from astropy.cosmology import FlatLambdaCDM
cosmo=FlatLambdaCDM(70, 0.3)

import iqbal_agn_heating_functions as iq
from cluster_measurements import ClusterMeasurements
from iq_agn_heating_2 import NFWProfile
```

```python
rs = np.logspace(-2, 1)*u.Mpc
```

```python
M2e14 = NFWProfile(2e14*u.Msun, 0)
print(M2e14.R500, M2e14.Rvir)
```

```python
plt.loglog(rs, M2e14.M_enc(rs)/(4./3.*np.pi*rs**3), label='avg rho within r')
plt.plot([1e-2, 1e1], [500*M2e14.rho_c().to(u.Msun/u.Mpc**3).value, 500*M2e14.rho_c().to(u.Msun/u.Mpc**3).value], label="rho_tot = 500*rho_c")

plt.scatter(M2e14.R500, 500*M2e14.rho_c().to(u.Msun/u.Mpc**3).value, alpha=0.3, label="R500")
plt.xlabel('r (Mpc)')
plt.ylabel('rho(Msun/Mpc3)')
plt.legend()
```
